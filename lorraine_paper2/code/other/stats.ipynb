```python
def get_means(model, source):
	df = data[(data['model'] == model) & (data['source'] == source)]

	if source != "inherent":
		print(f"means for {model} {source}:")
		print(f"\t{model} liberal ingroup: {df[(df['category'] == 'ingroup') & (df['alignment'] == 'liberal')]["response"].mean()}")
		print(f"\t{model} liberal fairness: {df[(df['category'] == 'fairness') & (df['alignment'] == 'liberal')]["response"].mean()}")
		print(f"\t{model} liberal purity: {df[(df['category'] == 'purity') & (df['alignment'] == 'liberal')]["response"].mean()}")
		print(f"\t{model} liberal authority: {df[(df['category'] == 'authority') & (df['alignment'] == 'liberal')]["response"].mean()}")
		print(f"\t{model} liberal harm: {df[(df['category'] == 'harm') & (df['alignment'] == 'liberal')]["response"].mean()}")
		print()

		print(f"\t{model} conservative ingroup: {df[(df['category'] == 'ingroup') & (df['alignment'] == 'conservative')]["response"].mean()}")
		print(f"\t{model} conservative fairness: {df[(df['category'] == 'fairness') & (df['alignment'] == 'conservative')]["response"].mean()}")
		print(f"\t{model} conservative purity: {df[(df['category'] == 'purity') & (df['alignment'] == 'conservative')]["response"].mean()}")
		print(f"\t{model} conservative authority: {df[(df['category'] == 'authority') & (df['alignment'] == 'conservative')]["response"].mean()}")
		print(f"\t{model} conservative harm: {df[(df['category'] == 'harm') & (df['alignment'] == 'conservative')]["response"].mean()}")
		print()
		print()
	else:
		print(f"means for {model} {source}:")
		print(f"\t{model} ingroup: {df[(df['category'] == 'ingroup')]["response"].mean()}")
		print(f"\t{model} fairness: {df[(df['category'] == 'fairness')]["response"].mean()}")
		print(f"\t{model} purity: {df[(df['category'] == 'purity') ]["response"].mean()}")
		print(f"\t{model}  authority: {df[(df['category'] == 'authority')]["response"].mean()}")
		print(f"\t{model} con harm: {df[(df['category'] == 'harm') ]["response"].mean()}")
		print()
		print()
```

```python
for model in data.model.unique().tolist():
    for source in data.source.unique().tolist():
        if (model == "homo sapiens neurons") & (source != "human"):
                print(end='')
        elif(model != "homo sapiens neurons") & (source == "human"):
                print(end='')
        else:
                get_means(model, source)
```
